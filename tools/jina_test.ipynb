```python
import requests
from datetime import datetime, timedelta

def get_yesterdays_fc_copenhagen_news(api_key):
    # Step 1: Calculate yesterday's date in ISO 8601 format
    today = datetime.now()
    formatted_today = today.strftime('%Y-%m-%d')
    yesterday = datetime.now() - timedelta(days=1)
    formatted_yesterday = yesterday.strftime('%Y-%m-%d')

    # Step 2: Set the URL for the NewsAPI's everything endpoint
    url = 'https://newsapi.org/v2/everything'

    # Step 3: Define the query parameters for F.C. København's news
    query_params = {
        'q': 'Arsenal',            # Exact match for 'F.C. København' using quotes
        'from': formatted_yesterday,        # Articles from the start of yesterday
        'to': formatted_yesterday,      # Until the end of yesterday
        #'domains' :'bold.dk',          
        'searchIn': 'title,description',    # Restrict search to titles and descriptions
        #'language': 'en',                   # Restrict articles to English language
        'sortBy': 'publishedAt',            # Sort articles by the published date
        'pageSize': 5,                     # Limit the number of articles to 10
        'apiKey': api_key                   # Your API key from NewsAPI
    }

    # Step 4: Make the request to NewsAPI
    response = requests.get(url, params=query_params)

    # Step 5: Check if the request was successful and return the news data
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to retrieve news. Status code: {response.status_code}")
        return None

# Example usage:
fc_copenhagen_news = get_yesterdays_fc_copenhagen_news(api_key)
```

```python
# contextual
import requests
from datetime import datetime, timedelta

def get_yesterdays_football_headlines(api_key):
    # Step 1: Calculate yesterday's date
    yesterday = datetime.now() - timedelta(days=1)
    formatted_yesterday = yesterday.strftime('%Y-%m-%dT00:00:00')
    today = datetime.now().strftime('%Y-%m-%dT23:59:59')
    
    # Step 2: Set the URL for the ContextualWeb News API
    url = "https://contextualwebsearch-websearch-v1.p.rapidapi.com/api/search/NewsSearchAPI"
    
    # Step 3: Define the query parameters for football news in Denmark
    query_params = {
        "q": "football",
        "fromPublishedDate": formatted_yesterday,
        "toPublishedDate": today,
        "pageSize": "10",  # Number of results
        "autoCorrect": "true",
        "safeSearch": "true",
        "withThumbnails": "true"
    }

    # Step 4: Set the headers for the request (including your API key)
    headers = {
        "X-RapidAPI-Key": api_key,
        "X-RapidAPI-Host": "contextualwebsearch-websearch-v1.p.rapidapi.com"
    }

    # Step 5: Make the request to ContextualWeb News API
    response = requests.get(url, headers=headers, params=query_params)
    
    # Step 6: Check if the request was successful and return the headlines
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to retrieve news. Status code: {response.status_code}")
        return None

# Example usage:
api_key = 'your_rapidapi_key'
football_headlines = get_yesterdays_football_headlines(api_key)

if football_headlines:
    for article in football_headlines['value']:
        print(f"Title: {article['title']}")
        print(f"Source: {article['provider']['name']}")
        print(f"Published at: {article['datePublished']}")
        print(f"URL: {article['url']}")
        print("\n")
```
